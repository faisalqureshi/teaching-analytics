# Course analysis

Python code snippet for plotting histograms of course grades and correlations between grades from two different exams.

```python
import sys
sys.path.append(".")
import plot

%matplotlib inline
```

```python
print(plot)
```

```python
df = plot.read_file('test1.xlsx', sheet_name=0)
```

## Using custom code that piggy backs on panda dataframe and matplotlib

```python
plot.inspect_columns(df)
```

Use column index to identify the columns that needs to be plotted (histogram).  You can also specify the total marks for each column.  This information is useful for computing percentage scores.

```python
c, t = plot.get_include_columns(df,cols='7,8',totals='100,90')
print(c)
print(t)
```

Go plot histogram.

```python
plot.plot_hist(df,c,t,'Test course')
```

To plot correlation, we need at least two columns.  

```python
plot.plot_corr(df,c,t,'Test course')
```

We can also plot letter grade histogram

```python
plot.plot_grades(df,c,t,'Test course')
```
